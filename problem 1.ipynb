```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn

%matplotlib inline
```

```python
data=pd.read_csv('problem_1/train.csv')
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
mean=data['Albumin_and_Globulin_Ratio'].mean()
```

```python
data.query("Albumin_and_Globulin_Ratio=='Nan'")
```

```python
data['Albumin_and_Globulin_Ratio'].fillna(mean,inplace=True)
```

```python
data.isnull().sum()
```

# Removing outliers

```python
from scipy import stats
num_df=data.select_dtypes(include=["number"])
idx=np.all(stats.zscore(num_df)<3,axis=1)
data[idx]
```

# Converting Gender into 0 or 1 values

```python
gen=list(data['Gender'])
lst=list()
for i in gen:
    if i=='Male':
        lst.append(1)
    else:
        lst.append(0)
data.drop(['Gender'], axis=1)
data['Gender']=lst
```

```python
data.head()
```

```python
X=data[['Age','Gender','Total_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase','Total_Protiens','Albumin','Albumin_and_Globulin_Ratio']]
y=data['Dataset']
```

# visualizing some data

```python
a, b = data['Dataset'].value_counts()
dct={
    "1":a,
    "2":b
}
dct
ticks = range(1,3)
height = list(dct.values())
tick_label = list(dct.keys())
plt.bar(ticks, height, tick_label=tick_label, color=['r', 'b'], width=0.9)
plt.show()
```

```python
corr = data.corr()
plt.figure(figsize=(30, 30))
import seaborn as sns
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
           cmap= 'coolwarm')
plt.title('Correlation between features');
```

# Classification using svm

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test= train_test_split(X,y,random_state=1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC 
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train) 
svm_predictions = svm_model_linear.predict(X_test)
```

```python
accuracy = svm_model_linear.score(X_test, y_test)
accuracy
```

# Classification using Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier 
X=data[['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase','Aspartate_Aminotransferase','Total_Protiens','Albumin','Albumin_and_Globulin_Ratio']]
X_train,X_test, y_train, y_test= train_test_split(X,y,random_state=1)
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train) 
dtree_predictions = dtree_model.predict(X_test) 
dtree_model.score(X_test,y_test)
```

# Classification using KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier 
X=data[['Age','Gender','Total_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase','Total_Protiens','Albumin','Albumin_and_Globulin_Ratio']]
X_train,X_test, y_train, y_test= train_test_split(X,y,random_state=1)
knn = KNeighborsClassifier(n_neighbors = 2).fit(X_train, y_train) 
accuracy = knn.score(X_test, y_test) 
accuracy
```

#  Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
X=data[['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase','Aspartate_Aminotransferase','Total_Protiens','Albumin','Albumin_and_Globulin_Ratio']]
X_train,X_test, y_train, y_test= train_test_split(X,y,random_state=1)
random_forest = RandomForestClassifier(n_estimators=61)
random_forest.fit(X_train, y_train)
rf_predicted = random_forest.predict(X_test)
random_forest.score(X_test,y_test)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
X=data[['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase','Aspartate_Aminotransferase','Total_Protiens','Albumin','Albumin_and_Globulin_Ratio']]
X_train,X_test, y_train, y_test= train_test_split(X,y,random_state=1)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
log_predicted= logreg.predict(X_test)
logreg.score(X_test,y_test)
```

# Naive Bayes classifier

```python
from sklearn.naive_bayes import GaussianNB
X=data[['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase','Alamine_Aminotransferase','Aspartate_Aminotransferase','Total_Protiens','Albumin','Albumin_and_Globulin_Ratio']]
X_train,X_test, y_train, y_test= train_test_split(X,y,random_state=1)
gnb = GaussianNB().fit(X_train, y_train) 
gnb_predictions = gnb.predict(X_test) 
gnb.score(X_test, y_test) 
```
